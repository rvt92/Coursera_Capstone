Part 1: import Wikipedia table and put it into a dataframe and clean the data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
dfs=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
dfs[0]
```

```python
#change column names to match the requirements
df=dfs[0]
df=df.rename(columns={'Postal Code':'PostalCode','Neighbourhood':'Neighborhood'})
df
```

Drop any rows with not assigned Borough

```python

df=df[df.Borough !='Not assigned']
df
```

Replace Not Assigned Neighborhoods with Borough

```python
# Replace any not assigned Neighborhood with its Borough
df.Neighborhood.replace('Not assigned',df.Borough,inplace=True)
df
```

PART 2

IMPORT csv with coordinates

```python
url = "https://cocl.us/Geospatial_data"
coordinates = pd.read_csv(url)

print(coordinates.head(10))
```

```python
#Rename columns to match the names
coordinates=coordinates.rename(columns={'Postal Code':'PostalCode'})
coordinates
#df= pd.merge(df,coordinates, on "")
```

Join both tables to get lat and long

```python
#merge dataframes to add latitude and longitude
df=pd.merge(df,coordinates,on='PostalCode',how='left')
df
```

PART 3: Select only Boroughs that include Toronto

```python
df=df[df['Borough'].str.contains("Toronto")]
df
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
